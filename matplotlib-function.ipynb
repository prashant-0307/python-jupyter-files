```python

# Data to plot
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [25, 30, 20, 25]  # Percent values
colors = ['red', 'yellow', 'pink', 'brown']
explode = (0, 0, 0, 0)  # Only "explode" the 2nd slice (Bananas)

# Create a pie chart
plt.pie(sizes, 
        labels=labels, 
        colors=colors, 
        explode=explode, 
        autopct='%1.1f%%', 
        #shadow=True, 
        startangle=90)

plt.title('Fruit Sales Distribution')
plt.axis('equal')  # Equal aspect ratio makes the pie a circle.
plt.show()
```
