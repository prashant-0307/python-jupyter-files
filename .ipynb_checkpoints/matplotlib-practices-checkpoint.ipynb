```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.randint(0,10,9)
y = np.random.randint(0,10,9)
print(x)
plt.plot(np.sort(x),y)
print(np.sort(x))
plt.title('sample line chart')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.figure(figsize=(8,5))
plt.plot(x, y, color='black', marker='o', linestyle='--', label='Line Plot')

plt.title('Simple Line Plot Example')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.savefig('simple_plot.png')
plt.show()
```
